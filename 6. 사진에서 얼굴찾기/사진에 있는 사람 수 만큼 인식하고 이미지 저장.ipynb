# 사진에 있는 사람 수 만큼 인식하고 이미지 저장

```python
from PIL import Image
import face_recognition
import datetime
import time
import cv2

# 이미지 로드
image = face_recognition.load_image_file("images\BTS.jpg")

# 기본 HOG 기반 모델을 사용하여 이미지의 모든 얼굴을 찾는다.
face_locations = face_recognition.face_locations(image)

print("I found {} face(s) in this photograph.".format(len(face_locations)))

for face_location in face_locations:
    
    # 이 이미지에서 각 얼굴의 위치를 출력한다.
    top, right, bottom, left = face_location
    print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

    face_image = image[top:bottom, left:right]
    face_image_RGB = cv2.cvtColor(face_image, cv2.COLOR_BGR2RGB)
    
    basename = "save_images\image"
    suffix = datetime.datetime.now().strftime("%y%m%d_%H%M%S") + '.jpg'
    filename = "_".join([basename, suffix])

    cv2.imwrite(filename, face_image_RGB)
    
    time.sleep(1)
```

# 인식 후 저장된 이미지

```python
import matplotlib.pyplot as plt
import cv2

fig = plt.figure(figsize=(16,16))

img1 = cv2.imread('save_images\image1.jpg')
img2 = cv2.imread('save_images\image2.jpg')
img3 = cv2.imread('save_images\image3.jpg')
img4 = cv2.imread('save_images\image4.jpg')
 
ax1 = fig.add_subplot(1, 4, 1)
ax1.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax1.axis("off")

ax2 = fig.add_subplot(1, 4, 2)
ax2.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax2.axis("off")

ax3 = fig.add_subplot(1, 4, 3)
ax3.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
ax3.axis("off")

ax4 = fig.add_subplot(1, 4, 4)
ax4.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
ax4.axis("off")

plt.show()
```
